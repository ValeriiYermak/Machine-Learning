```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import urllib.request
import missingno as msno
import os
os.environ["OMP_NUM_THREADS"] = "2"
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
url = 'https://raw.github.com/goitacademy/MACHINE-LEARNING-NEO/main/datasets/mod_05_topic_10_various_data.pkl'
with urllib.request.urlopen(url) as file:
    dataset = pickle.load(file)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
concrete = dataset['concrete']
concrete.info()
```

```python
concrete.head(10)
```

```python
feature_cols = concrete.columns.drop('CompressiveStrength')
concrete['Components'] = (concrete[feature_cols] > 0).sum(axis=1)
concrete.head(10)
```

```python
X = concrete.drop(columns=['CompressiveStrength'])
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = KMeans(random_state=37)
visualizer = KElbowVisualizer(model, k=(2,10))
visualizer.fit(X_scaled)
visualizer.show()
```

```python
optimal_k = visualizer.elbow_value_
kmeans = KMeans(n_clusters=optimal_k, random_state=37)
clusters = kmeans.fit_predict(X_scaled)
concrete['Cluster'] = clusters
```

```python
median_report = concrete.groupby('Cluster').median(numeric_only=True)

median_report.info()
```

```python
counts = concrete['Cluster'].value_counts().sort_index()
median_report['Recipes_count']=counts

print(median_report)
```

**Загальний висновок по завданню:**
* Кластеризація допомогла виділити типові рецепти бетонних сумішей.
* Є чіткий зв’язок між складом суміші та її міцністю на стискання.
* Для підвищення міцності рекомендовано використовувати FlyAsh (золу) та Superplasticizer (пластифікатор).
* median_report показує медіанні значення компонентів бетонної суміші для кожного кластера, включно з CompressiveStrength (міцність на стискання).
* Кожен кластер відображає типову «рецептуру» бетону.
* Кластер 0 характеризується заміною частини цементу на BlastFurnaceSlag (шлак). Дає помірну міцність (~39.7 МПа).
* Бачимо, що кластер 1 має найвищу медіанну міцність (~42.8) → рецепти з золою (FlyAsh) + пластифікатором дають кращу міцність.
* Кластер 2 — чисто цементна суміш без добавок, має нижчу міцність (~37.8).
* Кластер 3 — збалансований, але найнижча міцність (~33.3).
* Практичний висновок: Добавки (зола, шлак, пластифікатори) впливають на підвищення міцності.
